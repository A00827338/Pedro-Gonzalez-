```python
df = pd.read_csv("datasets/Cartwheeldata.csv")
df.head()
```

```python
sns.lmplot (x= "Wingspan" ,
            y= "CWDistance",
            data = df,
            fit_reg= False, #quitar la linea de regresi√≥n
            hue = "Gender" #identificar los puntos
           )
plt.title("Scatter plot of CWDistance vs Wingspan", fontsize=17)
plt.show()
```

```python
sns.swarmplot( x= "Gender",
              y="CWDistance",
              data= df
             )
plt.show()
```

```python
sns.boxplot(data=df.loc[:,["Age", "Height", "Wingspan","CWDistance","Score"]]) #grafica de boxplot
plt.xticks(rotation=45, fontsize=17)
plt.yticks(fontsize=17)
plt.title("Box plot representation of entire data", fontsize=15)
plt.show()
```

# Box plot of Male Data


```python
sns.boxplot(data=df.loc[df["Gender"]=="M",
                       ["Age", "Height", "Wingspan","CWDistance","Score"]])
                        
plt.xticks(rotation=45, fontsize=17)
plt.yticks(fontsize=17)
plt.title("Box plot representation of Male data", fontsize=15)
plt.show()
```

```python
fig= plt.figure()
ax1= fig.add_subplot(3,2,1) #of rows,#of columns position

sns.boxplot(data = df.loc[df["Gender"]=="M",
                         ["Score"]
                         ]
           )


ax2= fig.add_subplot(3,2,2) #of rows,#of columns position

sns.boxplot(data = df.loc[df["Gender"]=="F",
                         ["Score"]
                         ]
           )

ax3= fig.add_subplot(3,2,5) #of rows,#of columns position

sns.boxplot(data = df.loc[df["Gender"]=="M",
                         ["Score"]
                         ]
           )

ax4= fig.add_subplot(3,2,6) #of rows,#of columns position

sns.boxplot(data = df.loc[df["Gender"]=="F",
                         ["Score"]
                         ]
           )

ax1.title.set_text("Male box plot")
ax2.title.set_text("Female box plot")
ax3.title.set_text("Male box plot")
ax4.title.set_text("Female box plot")
plt.show()
           
```
