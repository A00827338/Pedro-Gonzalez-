```python
import seaborn as sns 
import matplotlib as mpl
import matplotlib.pyplot as plt 
import plotly.express as px 
```

```python
df = pd.read_excel("mic2.xlsx",
                   engine = "openpyxl")
df
```

```python
mpl.style.use("ggplot") #para usar tablas
```

```python
counts = df["Age"].value_counts()
```

```python
counts #contar la repeticion de cada condicion 
```

```python
counts= counts.reindex(["<=20","21 - 30","31 - 40","41 - 50"])
counts 
```

```python
type(counts)
counts = counts.to_frame()
# se cambio los datos de city a table, y ahora si puede leer los datos 
```

```python
x = counts.loc[:,'Age'].sum(axis=0)
x
```

```python
counts['%'] = (counts.iloc[:, :]/x)*100 #sacar porcentualmente los datos 
```

```python
counts
```

```python
counts.columns
```

```python
counts
```

```python
y = counts.reset_index()
y
```

```python
y.columns
```

```python
y = y.rename({'index':'Age_Group'},
                       axis = 1)
y
```

```python
x = list(y['Age_Group'])
y = list(y['Age'])

plt.subplot(1,2,1)
plt.rcParams["figure.figsize"]= (15,7)
plt.bar(x,y, width = 0.7, color = 'darkcyan')

explode_list = [0.08 , 0.09 , 0.07 , 0.06] # para hacer que se separen las partes del pie chart
color_list = [ 'gold' , 'yellowgreen', 'coral', 'skyblue', 'pink'] #cambiar colores de la piechart
y['Age'].plot(kind = 'pie',
                   figsize = (15,6),
                   autopct = '%1.1f%%', #es un numero 1, no una L 
                   startangle = 90,
                   shadow = True,
                   labels = None,
                   pctdistance = 1.1,
                   explode = explode_list, #llamar la explode list
                   colors = color_list, #llamar los colores 
)

                            
```
