```python
df = pd.read_csv("datasets/antropometria-dataset.csv")
df.head()
```

```python
df.describe()
```

```python
pesoxestado = pd.pivot_table(df,
              index ='desc_ent',
              values = 'peso'
                            )
pesoxestado
```

```python
mediana_sexo = pd.pivot_table(df,
                index ="desc_ent",
                values = "peso",
                aggfunc = "median" ,
                columns ="sexo",
                            )
mediana_sexo
```

```python
promedio_sexo = pd.pivot_table(df,
                index ="desc_ent",
                values = "peso",
                aggfunc = "mean" ,
                columns ="sexo",
                            )
promedio_sexo
```

```python
adf = pd.read_csv("datasets/adult.csv")
adf
```

```python
prom_horas = pd.pivot_table(adf,
                     index = "marital-status",
                     values = "hours-per-week",
                     aggfunc ="mean",
                     columns = "education",
                    )
prom_horas
```

```python
median_age = pd.pivot_table(adf,
                     index = "marital-status",
                     values = "age",
                     aggfunc ="median",
                     columns = ["education", "sex"],
                    )
median_age
```
